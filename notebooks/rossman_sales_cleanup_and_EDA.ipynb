## 1. Loading Dataset

#### Data Fields
**Data Source** : https://www.kaggle.com/c/rossmann-store-sales/data

- **Id** - An Id that represents a (Store, Date) duple within the test set
- **Store** - A unique Id for each store
- **Sales** - The turnover for any given day (this is what we are predicting)
- **Customers** - The number of customers on a given day
- **Open** - An indicator for whether the store was open: 0 = closed, 1 = open
- **StateHoliday** - Indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. Note that all schools are closed on public holidays and weekends. a = public holiday, b = Easter holiday, c = Christmas, 0 = None
- **SchoolHoliday** - Indicates if the (Store, Date) was affected by the closure of public schools
- **StoreType** - Differentiates between 4 different store models: a, b, c, d
- **Assortment** - Describes an assortment level: a = basic, b = extra, c = extended
- **CompetitionDistance** - Distance in meters to the nearest competitor store
- **CompetitionOpenSince[Month/Year]** - Gives the approximate year and month of the time the nearest competitor was opened
- **Promo** - Indicates whether a store is running a promo on that day
- **Promo2** - Promo2 is a continuing and consecutive promotion for some stores: 0 = store is not participating, 1 = store is participating
- **Promo2Since[Year/Week]** - Describes the year and calendar week when the store started participating in Promo2
- **PromoInterval** - Describes the consecutive intervals Promo2 is started, naming the months the promotion is started anew. E.g. "Feb,May,Aug,Nov" means each round starts in February, May, August, November of any given year for that store

## 2. Cleaning and Exploring Datasets

In this section, we verify, clean and explore the categorical and continuous features of the dataset. For continuous variables, we check for the basic assumption of linear regression models, such as linearity and no multicollinearity. For categorical variables, we generate dummies (one-hot encoding) .

#### Exploring features

- **Date**

```python
#Verifying date format
pd.to_datetime(combined_data['Date'], errors='coerce').isna().sum()
```

```python
#Checking date range
print('start date: ', combined_data['Date'].min())
print('end date: ', combined_data['Date'].max())
```

```python
#Splitting date into day (1-31), month (1-12) and year.
datetime_idx = pd.DatetimeIndex(combined_data['Date'])
combined_data['DayOfMonth'] = datetime_idx.day
combined_data['Month'] = datetime_idx.month
combined_data['Year'] = datetime_idx.year

# Plotting sales wrt day, month and year
fig, axes = plt.subplots(2,2, figsize=(8,6))
sns.barplot(x='DayOfWeek', y='Sales', data=combined_data, ax=axes[0,0], estimator='mean', palette='coolwarm', errorbar=None)
sns.barplot(x='DayOfMonth', y='Sales', data=combined_data, ax=axes[0,1], estimator='mean', palette='coolwarm', errorbar=None)
axes[0,1].set_xticks(range(1,31,2))
sns.barplot(x='Month', y='Sales', data=combined_data, ax=axes[1,0], estimator='mean', palette='coolwarm', errorbar=None)
sns.barplot(x='Year', y='Sales', data=combined_data, ax=axes[1,1], estimator='mean', palette='coolwarm', errorbar=None)
fig.suptitle('Avg. Sales (/store/day)')
plt.tight_layout()
plt.show()
```

```python
# Generate dummies (one-hot encoding) for day of week
encoded_DayOfWeek = pd.get_dummies(combined_data['DayOfWeek'], dtype=np.uint8)
week_mapping = {1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat', 7: 'Sun'}
encoded_DayOfWeek.rename(columns=week_mapping, inplace=True)

encoded_DayOfWeek.drop_duplicates().head()

#One-hot encoding using get_dummies can introduce multicollinearity issue.
#If we have a categorical variable with three values, then we create three one-hot encoded variables (X, Y, Z). 
#Since, X+Y+Z=1, one of the variables can always be obtained in terms of other.
#Solution: Drop one of the one-hot encoded variable.
```

```python
#one-hot encoding: Month
encoded_Month = pd.get_dummies(combined_data['Month'], dtype=np.uint8)
month_mapping = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 
                 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
encoded_Month.rename(columns=month_mapping, inplace=True)

encoded_Month.drop_duplicates().head()
```

```python
#one-hot encoding: Year
encoded_Year = pd.get_dummies(combined_data['Year'], dtype=np.uint8)

encoded_Year.drop_duplicates().head()
```

```python
#one-hot encoding: StateHoliday
encoded_StateHoliday = pd.get_dummies(combined_data['StateHoliday'], dtype=np.uint8)
stateholiday_mapping = {'0': 'H_None', 'a': 'H_Public', 'b': 'H_Easter', 'c': 'H_XMas'}
encoded_StateHoliday.rename(columns=stateholiday_mapping, inplace=True)

encoded_StateHoliday.drop_duplicates().head()
```

```python
#Label Encoding: StateHoliday
stateholiday_num_mapping = {'0': 0, 'a': 1, 'b': 2, 'c': 3}
combined_data['StateHoliday'] = combined_data['StateHoliday'].map(stateholiday_num_mapping)
```

- **Store type and Assortment**

```python
fig, axes = plt.subplots(1,2, figsize=(8,3))
sns.barplot(x='StoreType', y='Sales', data=combined_data, ax = axes[0], estimator='mean', palette='coolwarm', errorbar=None)
sns.barplot(x='Assortment', y='Sales', data=combined_data, ax = axes[1], estimator='mean', palette='coolwarm', errorbar=None)
fig.suptitle('Avg. Sales (/store/day)')
plt.tight_layout()
plt.show()
```

```python
#one-hot encoding: StoreType
encoded_StoreType = pd.get_dummies(combined_data['StoreType'], dtype=np.uint8)
storetype_mapping = {'a': 'StoreType_A', 'b': 'StoreType_B', 'c': 'StoreType_C', 'd': 'StoreType_D'}
encoded_StoreType.rename(columns=storetype_mapping, inplace=True)

encoded_StoreType.drop_duplicates().head()
```

```python
#Label Encoding: StoreType
StoreType_num_mapping = {'a': 0, 'b': 1, 'c': 2, 'd': 3}
combined_data['StoreType'] = combined_data['StoreType'].map(StoreType_num_mapping)
```

```python
#one-hot encoding: Assortment
encoded_Assortment = pd.get_dummies(combined_data['Assortment'], dtype=np.uint8)
assort_mapping = {'a': 'Assort_Basic', 'b': 'Assort_Extra', 'c': 'Assort_Extended'}
encoded_Assortment.rename(columns=assort_mapping, inplace=True)

encoded_Assortment.drop_duplicates().head()
```

```python
#Label Encoding: Assortment
Assortment_num_mapping = {'a': 0, 'b': 1, 'c': 2}
combined_data['Assortment'] = combined_data['Assortment'].map(Assortment_num_mapping)
```

- **Competition**: Let's look closely at variables related to competitors.

We derive new features: 
1. *Log_CompetitionYears*: The log of number of years the nearest competitor was opened.
2. *Log_CompetitionDistance*: The log of distance in kilometers to the nearest competitor store

```python
combined_data['Log_CompetitionYears'] = ((combined_data['Year']-combined_data['CompetitionOpenSinceYear'])*12 + (combined_data['Month']-combined_data['CompetitionOpenSinceMonth']))/12
combined_data['Log_CompetitionYears'] = combined_data['Log_CompetitionYears'].apply(np.log)
combined_data['Log_CompetitionDistance'] = combined_data['CompetitionDistance'].apply(np.log)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,3))
sns.scatterplot(x='Log_CompetitionYears', y='Sales', data=combined_data, ax = axes[0], alpha=0.1)
sns.scatterplot(x='Log_CompetitionDistance', y='Sales', data=combined_data, ax = axes[1], alpha=0.1)
fig.suptitle('Avg. Sales (/store/day)')
plt.tight_layout()
plt.show()
```

- **Promo**: Let's explore variables related to promotions.

We derive a new feature: 
1. *Promo2Years*: The number of years since the store started participating in Promo2. A positive value means the store started participating in promo before the given sales date. A negative value implies the store starts participating in promo after the given sales date.

```python
combined_data['Promo2Years'] = ((combined_data['Year']-combined_data['Promo2SinceYear'])*52 + (combined_data['Month']*(52/12)-combined_data['Promo2SinceWeek']))/52
combined_data['Promo2Years'] = combined_data['Promo2Years'].round(2)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,3))
sns.scatterplot(x='Promo2Years', y='Sales', data=combined_data, ax = axes[0], alpha=0.1)
sns.barplot(x='PromoInterval', y='Sales', data=combined_data, ax = axes[1], estimator='mean', palette='coolwarm', errorbar=None)
fig.suptitle('Avg. Sales (/store/day)')
plt.tight_layout()
plt.show()
```

```python
#one-hot encoding: PromoInterval
encoded_PromoInterval = pd.get_dummies(combined_data['PromoInterval'], dtype=np.uint8)
PromoInterval_mapping = {'Feb,May,Aug,Nov': 'Promo2_Feb,May,Aug,Nov', 
                     'Jan,Apr,Jul,Oct': 'Promo2_Jan,Apr,Jul,Oct', 
                     'Mar,Jun,Sept,Dec': 'Promo2_Mar,Jun,Sept,Dec'}
encoded_PromoInterval.rename(columns=PromoInterval_mapping, inplace=True)

encoded_PromoInterval.drop_duplicates().head()
```

```python
#Label Encoding: PromoInterval
PromoInterval_num_mapping = {'Feb,May,Aug,Nov': 0, 
                             'Jan,Apr,Jul,Oct': 1, 
                             'Mar,Jun,Sept,Dec': 2}
combined_data['PromoInterval'] = combined_data['PromoInterval'].map(PromoInterval_num_mapping)
```

#### Updating dataset

```python
#Dropping old variables
combined_data.drop(columns=['CompetitionDistance'], inplace=True)
combined_data.drop(columns=['CompetitionOpenSinceMonth'], inplace=True)
combined_data.drop(columns=['CompetitionOpenSinceYear'], inplace=True)
combined_data.drop(columns=['Promo2SinceWeek'], inplace=True)
combined_data.drop(columns=['Promo2SinceYear'], inplace=True)


# Add dummies (encoded variables) to our dataset
combined_data = pd.concat([combined_data, encoded_DayOfWeek], axis=1)
combined_data = pd.concat([combined_data, encoded_Month], axis=1)
combined_data = pd.concat([combined_data, encoded_Year], axis=1)
combined_data = pd.concat([combined_data, encoded_StateHoliday], axis=1)
combined_data = pd.concat([combined_data, encoded_StoreType], axis=1)
combined_data = pd.concat([combined_data, encoded_Assortment], axis=1)
combined_data = pd.concat([combined_data, encoded_PromoInterval], axis=1)
```

#### Filling missing values in the dataset


```python
# Checking missing values in the dataset
print("Total size of dataset: ", len(combined_data))
null_counts = combined_data.isnull().sum()
print(null_counts[null_counts>0])
```

- The missing values in *Log_CompetitionDistance*, *Log_CompetitionYears* are filled using the mean value.
- The missing values in *Promo2Years*, *PromoInterval* are filled as 0 (based on value of *Promo2*)


```python
from sklearn.impute import SimpleImputer

combined_data['Log_CompetitionYears'] = combined_data['Log_CompetitionYears'].replace([np.inf, -np.inf], np.nan)

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
X = imp_mean.fit_transform(combined_data[['Log_CompetitionDistance', 'Log_CompetitionYears']])
combined_data['Log_CompetitionDistance'] = X[:,0]
combined_data['Log_CompetitionYears'] = X[:,1]
```

```python
#Checking if Promo2 = 0 matches with NaN in Promo2SinceWeek, Promo2SinceYear, PromoInterval
X1 = (combined_data['Promo2'] == 0)
print(np.all(X1 == combined_data['Promo2Years'].isna()))
print(np.all(X1 == combined_data['PromoInterval'].isna()))
```

```python
combined_data['Promo2Years'] = combined_data['Promo2Years'].fillna(0)
combined_data['PromoInterval'] = combined_data['PromoInterval'].fillna(0)
```
